```python
def fib(num):
    if (num==1 or num == 2):    # that's my base case
        return 1
    fib_num_1 = fib(num-1)
    fib_num_2 = fib(num-2)
    output = fib_num_1 + fib_num_2
    return output
```

```python
def like(num):
    if (num==4):
        return num
    else:
        return 2*like(num+1)
```

```python
def isSorted(a):
    l = len(a)
    if l==0 or l==1:
        return True
    if a[0]>a[1]:
        return False
    smallerList = a[1:]
    isSmallerListSorted = isSorted(smallerList)
    # return isSmallerListSorted
    if isSmallerListSorted:
        return True
    else:
        return False
```

```python
def array_sum(a,length):
    if (len(a) == 0):
        return 0
    if (len(a) == 1):
        return a[0]
    return a[0]+array_sum(a[1:],length)
```

```python
def check(a,b):
    if len(a)==0:
        return "false"
    if b==a[0]:
        return "true"
    return check(a[1:],b)
```

```python
def checksorted(a,si):
    l = len(a)
    if si== l or si== l-1:    #si = start index
        return True
    if a[si] > a[si+1]:
        return False
    else:
        return checksorted(a,si+1)
```

```python
def firstoccurance(a,b,si):
    l = len(a)
    if a[si]!=b and si==l-1:
        return False
    if a[si]==b:
        return si
    else:
        return firstoccurance(a,b,si+1)
```

```python
num = 2
print(fib(num))
print(like(num))
a = [8,2,3,4,5]
b = 1
length = len(a)
print(isSorted(a))
print(array_sum(a,length))
check(a,b)
checksorted(a,0)
firstoccurance(a,b,0)
```
